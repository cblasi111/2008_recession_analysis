```python
# Converting 2007 GDP
convert_2007 = []

for gdp in gdp_df["2007"]:
    if gdp=="(L)":
        convert_2007.append(0)
    else:
        convert_2007.append(float(gdp))
        
    
gdp_df["new_2007"]=convert_2007

# Grouping by state and sum it by state
grouped_gdp_2007 = gdp_df.groupby(["GeoName"])["new_2007"].sum()
grouped_gdp_2007
```
