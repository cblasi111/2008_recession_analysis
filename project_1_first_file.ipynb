```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
pop_file = "Resources/Population Estimates.csv"
pop_file_df = pd.read_csv(pop_file, skiprows=[0])
pop_file_df.head()
```

```python
pop_df = pop_file_df.set_index("These data were posted to the ERS website (at https://www.ers.usda.gov/data-products/county-level-data-sets/download-data.aspx) May 2019. Contact: Tim Parker, ERS/USDA, timothy.parker@usda.gov, (202) 694-5435.")
pop_df.head()
```

```python
pop_df = pop_df.dropna(how='all')
pop_df.head()
```

```python
somerset_county = pop_df.loc["34035", "Unnamed: 64"]
print("Somerset County's Net domestic migration period from 4/1/2010 to 6/30/2010 is " + somerset_county + " people.")

monmouth_county = pop_df.loc["34025", "Unnamed: 64"]
print("Monmouth County's Net domestic migration period from 4/1/2010 to 6/30/2010 is " + monmouth_county + " people.")

essex_county = pop_df.loc["34013", "Unnamed: 64"]
print("Essex County's Net domestic migration period from 4/1/2010 to 6/30/2010 is " + essex_county + " people.")

cape_may_county = pop_df.loc["34009", "Unnamed: 64"]
print("Cape May County's Net domestic migration period from 4/1/2010 to 6/30/2010 is " + cape_may_county + " people.")

bergen_county = pop_df.loc["34003", "Unnamed: 64"]
print("Bergen County's Net domestic migration period from 4/1/2010 to 6/30/2010 is " + bergen_county + " people.")
```

```python
nj_top_pop_stats = pop_df.loc[["34035", "34025", "34013", "34009", "34003"], ["Unnamed: 64", "Unnamed: 19", "Unnamed: 8"]]
# 34035 = somerset, 34025 = monmouth, 34013 = essex, 34009 = capemay, 34003 = bergen
# Unnamed 64 = net domestic migration, unnamed 19 = , unnamed 8 = total resident population
print(nj_top_pop_stats)
print()
```

```python
nj_total_pop_stats = pop_df.loc[["34000"], ["Unnamed: 64", "Unnamed: 19", "Unnamed: 8"]]
# 34000 = NJ
# Unnamed 64 = net domestic migration, unnamed 19 = , unnamed 8 = total resident population
print(nj_total_pop_stats)
print()
```
