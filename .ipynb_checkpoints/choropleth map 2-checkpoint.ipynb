```python
# import dependencies and setup enviornment to work offline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
```

```python
# use init_notebook_mode to connect the Js to display the plots in the notebook
init_notebook_mode(connected=True)
```

```python
pop_file = "Resources/Population Estimates.csv"
pop_df = pd.read_csv(pop_file, skiprows=[0])
pop_df.head()
```

```python
new_header = pop_df.iloc[0]
pop_df = pop_df[1:]
pop_df.columns = new_header
pop_df.head(50)
```

```python
pop_df = pop_df.drop_duplicates(subset='State', keep='first', inplace=False)
pop_df.head(50)
```

```python
pop_df = pop_df.drop(["Rural-urban_Continuum Code_2003", "Rural-urban_Continuum Code_2013", "Urban_Influence_Code_2003", "Urban_Influence_Code_2013", "Economic_typology_2015"], axis=1)
pop_df.head(53)
```

```python
pop_df_clean = pop_df.drop([1])
pop_df_clean.head(50)
```

```python
# create a data object with the dict() method

# colorscale = 'Earth' --> this is the color for the geographical map elements
# locations = pop_df['Unnamed: 0'] --> this is the index for the states
# locationmode = 'USA-states' --> this will tell plotly we want the US
# z = pop_df['census2010pop'] --> numerical value for each state element
# text = pop_df['state'] --> this is the categorical value for each element
# colorbar = ['title': '2010 Census Population'] --> title on the right

data = dict(type = 'choropleth',
           colorscale = 'Jet',
           locations = pop_df_clean['State'],
           locationmode = 'USA-states',
           z = pop_df_clean['R_DOMESTIC_MIG_2011'],
           text = pop_df_clean['Area_Name'],
           colorbar = {'title': 'Migration Rate'},
           marker = dict(line = dict(color = 'rgb(255, 255, 255)', width=2)))
```

```python
# create a layout object with the dict() method

# scope arguemnts to the map of interest
# showlakes argument turns on display for lakes
# lakecolor adjusts color of lakes

layout = dict(title = 'Net Domestic Migration Rate 2010-2011',
             geo = dict(scope = 'usa', showlakes=True, lakecolor='rgb(85,173,240)'))
```

```python
# use the go.figure to set up the object that gets passed into iplot() method

choromap = go.Figure(data = [data], layout = layout)
```

```python
# use iplot() method to generate the plot

iplot(choromap)
```
