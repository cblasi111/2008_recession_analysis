```python
#imports csv into pandas, skips first 4 rows
poverty2 = "Resources/Poverty/hstpov21.csv"
poverty2_pd = pd.read_csv(poverty2, skiprows=4)
poverty2_df = pd.DataFrame(poverty2_pd)
poverty2_df.head()
```

```python
#erge_table = pd.join(pov2017, pov2016, pov2015, pov2014, pov2013, on="STATE", how="outer")
#merge_table = pd.merge(pov2017, pov2016, join="outer")
merge_table = pd.merge(pov2007_reset_renamed_col, pov2008_reset_renamed_col, on="STATE", how="outer")
merge_table
merge_table2 = pd.merge(merge_table, pov2017_reset_renamed_col, on="STATE", how="outer")
#print(merge_table2)
```

```python
##changes Percent_2007 to float
convert_2007 = []

for x in merge_table2["Percent_2007"]:
    convert_2007.append(float(x))

merge_table2["New percent 2007"] = convert_2007
print(merge_table2)
merge_table2.dtypes
```

```python
##changes Percent_2008 to float
convert_2008 = []

for x in merge_table2["Percent_2008"]:
    convert_2008.append(float(x))

merge_table2["New percent 2008"] = convert_2008
print(merge_table2)
```

```python
##changes Percent_2017 to float
convert_2017 = []

for x in merge_table2["Percent_2017"]:
    convert_2017.append(float(x))

merge_table2["New percent 2017"] = convert_2017
#print(merge_table2)
merge_table2.dtypes
```

```python
#table for visualization
table_vis = merge_table2[['STATE', 'Total_2007', 'Number_2007', 'New percent 2007', 'Total_2008',
       'Number_2008', 'New percent 2008', 'Total_2017', 'Number_2017', 'New percent 2017']]
table_vis
```

```python
delta_07_08 = table_vis["New percent 2008"] - table_vis["New percent 2007"]
delta_07_08
merge_table2["Change in Percent from 2007 to 2008"] = delta_07_08
merge_table2
```

```python
delta_08_17 = table_vis["New percent 2017"] - table_vis["New percent 2008"]
delta_08_17
merge_table2["Change in Percent from 2008 to 2017"] = delta_08_17
merge_table2
```

```python
#merge_deltas = pd.merge(delta_07_08, delta_08_17, on="STATE", how="outer")
poverty_changes_for_vis = merge_table2[["STATE", "Change in Percent from 2008 to 2017", "Change in Percent from 2007 to 2008"]]
poverty_changes_for_vis.head()
```

```python
state_initials = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "DoC","FL",
                 "GA","HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA","ME",
                 "MD","MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV","NH",
                 "NJ","NM", "NY", "NC", "ND", "OH", "OK", "OR","PA","RI",
                 "SC", "SD", "TN", "TX", "UT", "VT", "VA", "WA","WV","WI", "WY"]
```

```python
poverty_changes_for_vis["Abbreviations"] = state_initials
poverty_changes_for_vis
```

```python
# create a data object with the dict() method

# colorscale = 'Earth' --> this is the color for the geographical map elements
# locations = pop_df['Unnamed: 0'] --> this is the index for the states
# locationmode = 'USA-states' --> this will tell plotly we want the US
# z = pop_df['census2010pop'] --> numerical value for each state element
# text = pop_df['state'] --> this is the categorical value for each element
# colorbar = ['title': '2010 Census Population'] --> title on the right

data = dict(type = 'choropleth',
          colorscale = 'Earth',
          locations = poverty_changes_for_vis['Abbreviations'],  #abbreviations
          locationmode = 'USA-states',
          z = poverty_changes_for_vis['Change in Percent from 2008 to 2017'],
          text = poverty_changes_for_vis['STATE'],
          colorbar = {'title': 'Change in % Poverty'},
          marker = dict(line = dict(color = 'rgb(255, 255, 255)', width=2)))

# create a layout object with the dict() method

# scope arguemnts to the map of interest
# showlakes argument turns on display for lakes
# lakecolor adjusts color of lakes

layout = dict(title = 'Change in Poverty Percentage from 2008 to 2017',
            geo = dict(scope = 'usa', showlakes=True, lakecolor='rgb(85,173,240)'))
# use the go.figure to set up the object that gets passed into iplot() method

choromap = go.Figure(data = [data], layout = layout)
# use iplot() method to generate the plot

iplot(choromap)
```
