```python
gdp_complete = pd.DataFrame({"GDP 2007": grouped_gdp_2007, 
                             "GDP 2008": grouped_gdp_2008, 
                             "GDP 2009": grouped_gdp_2009, 
                             "GDP 2010": grouped_gdp_2010, 
                             "GDP 2011": grouped_gdp_2011, 
                             "GDP 2012": grouped_gdp_2012, 
                             "GDP 2013": grouped_gdp_2013, 
                             "GDP 2014": grouped_gdp_2014, 
                             "GDP 2015": grouped_gdp_2015, 
                             "GDP 2016": grouped_gdp_2016, 
                             "GDP 2017": grouped_gdp_2017, 
                             "GDP 2018": grouped_gdp_2018, 
                             "07-08 Delta": delta_07_08, 
                             "08-09 Delta": delta_08_09, 
                             "09-10 Delta": delta_09_10, 
                             "10-11 Delta": delta_10_11, 
                             "11-12 Delta": delta_11_12, 
                             "12-13 Delta": delta_12_13, 
                             "13-14 Delta": delta_13_14, 
                             "14-15 Delta": delta_14_15, 
                             "15-16 Delta": delta_15_16, 
                             "16-17 Delta": delta_16_17, 
                             "17-18 Delta": delta_17_18, 
                             "Total Delta": total_delta})

#gdp_complete.drop(["Far West", "Great Lakes", "Mideast", "New England", "Plains", 
                   #"Rocky Mountain", "Southeast", "Southwest"])
gdp_complete = gdp_complete.reset_index()
gdp_complete
```

```python
# Create dataframe to show 07-08 change and 08-18 change
gdp_change_07_08 = gdp_complete["09-10 Delta"]
gdp_change_08_18 = ((gdp_complete["GDP 2018"] - gdp_complete["GDP 2008"])/ gdp_complete["GDP 2008"])*100
state_name = [state for state in gdp_complete["GeoName"]]

state_initials = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "DoC","FL",
                  "GA","HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA","ME",
                  "MD","MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV","NH",
                  "NJ","NM", "NY", "NC", "ND", "OH", "OK", "OR","PA","RI",
                  "SC", "SD", "TN", "TX", "UT", "VT", "VA", "WA","WV","WI", "WY"]

gdp_for_visualization = pd.DataFrame({"State": state_name,
                                      #"Abbreviation": state_initials,
                                      "GDP Change (2007-2008)": gdp_change_07_08, 
                                      "GDP Change (2008-2018)": gdp_change_08_18})

gdp_for_visualization = gdp_for_visualization.set_index("State")
gdp_for_visualization = gdp_for_visualization.drop("Far West", axis=0)
gdp_for_visualization = gdp_for_visualization.drop("Great Lakes", axis=0)
gdp_for_visualization = gdp_for_visualization.drop("Mideast", axis=0)
gdp_for_visualization = gdp_for_visualization.drop("New England", axis=0)
gdp_for_visualization = gdp_for_visualization.drop("Plains", axis=0)
gdp_for_visualization = gdp_for_visualization.drop("Rocky Mountain", axis=0)
gdp_for_visualization = gdp_for_visualization.drop("Southeast", axis=0)
gdp_for_visualization = gdp_for_visualization.drop("Southwest", axis=0)
gdp_for_visualization = gdp_for_visualization.drop("United States *", axis=0)
gdp_for_visualization = gdp_for_visualization.reset_index()

gdp_for_visualization["Abbreviation"] = state_initials

gdp_for_visualization
```

```python
top_states_07_08 = gdp_for_visualization.sort_values(["GDP Change (2007-2008)"],ascending = False).head(5)
del top_states_07_08['GDP Change (2008-2018)']
del top_states_07_08['Abbreviation']

bottom_states_07_08 = gdp_for_visualization.sort_values(["GDP Change (2007-2008)"],ascending = True).head(5)
del bottom_states_07_08['GDP Change (2008-2018)']
del bottom_states_07_08['Abbreviation']
```

```python
# create a data object with the dict() method

# colorscale = 'Earth' --> this is the color for the geographical map elements
# locations = gdp_for_visualization['Unnamed: 0'] --> this is the index for the states
# locationmode = 'USA-states' --> this will tell plotly we want the US
# z = gdp_for_visualization['GDP Change (2007-2008)'] --> numerical value for each state element
# text = gdp_for_visualization['state'] --> this is the categorical value for each element
# colorbar = ['title': '2010 Census Population'] --> title on the right

data = dict(type = 'choropleth',
           colorscale = 'YlOrRd',
           locations = gdp_for_visualization["Abbreviation"],
           locationmode = 'USA-states',
           z = gdp_for_visualization['GDP Change (2007-2008)'],
           text = gdp_for_visualization["State"],
           colorbar = {'title': 'Change in GDP'},
           marker = dict(line = dict(color = 'rgb(255, 255, 255)', width=2))) 
```

```python
# create a layout object with the dict() method

# scope arguemnts to the map of interest
# showlakes argument turns on display for lakes
# lakecolor adjusts color of lakes

layout = dict(title = 'Change in GDP from 2007 to 2008 (%)',
             geo = dict(scope = 'usa', showlakes=True, lakecolor='rgb(85,173,240)'))
```

```python
# use the go.figure to set up the object that gets passed into iplot() method
choromap = go.Figure(data = [data], layout = layout)
```

```python
# use iplot() method to generate the plot
print(top_states_07_08)
print(bottom_states_07_08)
iplot(choromap)
```
