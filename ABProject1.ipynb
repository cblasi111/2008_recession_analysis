```python
# import dependencies and setup enviornment to work offline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot

economic_unemployed = "Resources/Unemployment and Median Income by state and year.csv"
```

```python
init_notebook_mode(connected=True)
```

```python
economic_unemployed_df = pd.read_csv(economic_unemployed, delimiter =",", encoding="utf-8-sig", skiprows=7)
economic_unemployed_df.columns
```

```python
economic_data = economic_unemployed_df.rename(columns={"Unemployment_rate_2007":"Unemployment rate 2007",
                                               "Unemployment_rate_2008":"Unemployment rate 2008",
                                                "Unemployment_rate_2009":"Unemployment rate 2009", 
                                                "Unemployment_rate_2010":"Unemployment rate 2010",
                                               "Unemployment_rate_2011":"Unemployment rate 2011",
                                               "Unemployment_rate_2012":"Unemployment rate 2012",
                                               "Unemployment_rate_2013":"Unemployment rate 2013",
                                               "Unemployment_rate_2014":"Unemployment rate 2014",
                                               "Unemployment_rate_2015":"Unemployment rate 2015",
                                               "Unemployment_rate_2016":"Unemployment rate 2016",
                                               "Unemployment_rate_2017":"Unemployment rate 2017",
                                               "Unemployment_rate_2018":"Unemployment rate 2018"})
economic_data.columns
```

```python
#data from 2007 to 2010
economic_unemployed_table = economic_data[["State", "Area_name", 'Unemployment rate 2007','Unemployment rate 2008','Unemployment rate 2009',"Unemployment rate 2010","Unemployment rate 2011","Unemployment rate 2012","Unemployment rate 2013","Unemployment rate 2014","Unemployment rate 2015","Unemployment rate 2016","Unemployment rate 2017","Unemployment rate 2018"]]
economic_unemployed_table
```

```python
#Create Unemployed rate table per each state
economic_df=economic_unemployed_table
economic_df
#extract just states based off column name Area_name
economic_df_states = economic_df[~economic_df['Area_name'].str.contains(",")]
```

```python
economic_df_states
```

```python
rate_07_08 = (economic_df_states["Unemployment rate 2008"] - economic_df_states["Unemployment rate 2007"])
rate_08_18 = (economic_df_states["Unemployment rate 2018"] - economic_df_states["Unemployment rate 2008"])
```

```python
rate_07_08
```

```python
rate_08_18
```

```python
state_table = pd.DataFrame({"Unemployment Rate Change from 2007 to 2008":rate_07_08, "Unemployment Rate Change from 2008 to 2018": rate_08_18})
```

```python
state_table
```

```python
economic_df_states["Rate Change from 2007 to 2008"]=state_table['Unemployment Rate Change from 2007 to 2008']
economic_df_states["Rate Change from 2008 to 2018"]=state_table['Unemployment Rate Change from 2008 to 2018']
```

```python
economic_df_states
```

```python
top_states_unemployed = economic_df_states.sort_values(['Rate Change from 2007 to 2008'],ascending = False).head(5)
del top_states_unemployed["Rate Change from 2008 to 2018"]
del top_states_unemployed["State"]
del top_states_unemployed["Unemployment rate 2007"]
del top_states_unemployed["Unemployment rate 2008"]
del top_states_unemployed["Unemployment rate 2009"]
del top_states_unemployed["Unemployment rate 2010"]
del top_states_unemployed["Unemployment rate 2011"]
del top_states_unemployed["Unemployment rate 2012"]
del top_states_unemployed["Unemployment rate 2013"]
del top_states_unemployed["Unemployment rate 2014"]
del top_states_unemployed["Unemployment rate 2015"]
del top_states_unemployed["Unemployment rate 2016"]
del top_states_unemployed["Unemployment rate 2017"]
del top_states_unemployed["Unemployment rate 2018"]



bottom_states_unemployed = economic_df_states.sort_values(['Rate Change from 2007 to 2008'],ascending = True).head(5)
del bottom_states_unemployed["Rate Change from 2008 to 2018"]
del bottom_states_unemployed["State"]
del bottom_states_unemployed["Unemployment rate 2007"]
del bottom_states_unemployed["Unemployment rate 2008"]
del bottom_states_unemployed["Unemployment rate 2009"]
del bottom_states_unemployed["Unemployment rate 2010"]
del bottom_states_unemployed["Unemployment rate 2011"]
del bottom_states_unemployed["Unemployment rate 2012"]
del bottom_states_unemployed["Unemployment rate 2013"]
del bottom_states_unemployed["Unemployment rate 2014"]
del bottom_states_unemployed["Unemployment rate 2015"]
del bottom_states_unemployed["Unemployment rate 2016"]
del bottom_states_unemployed["Unemployment rate 2017"]
del bottom_states_unemployed["Unemployment rate 2018"]
```

```python
# create a data object with the dict() method

# colorscale = 'Earth' --> this is the color for the geographical map elements
# locations = pop_df['Unnamed: 0'] --> this is the index for the states
# locationmode = 'USA-states' --> this will tell plotly we want the US
# z = pop_df['census2010pop'] --> numerical value for each state element
# text = pop_df['state'] --> this is the categorical value for each element
# colorbar = ['title': '2010 Census Population'] --> title on the right
data = dict(type = 'choropleth',
          colorscale = 'Jet',
          locations = economic_df_states['State'],
          locationmode = 'USA-states',
          z = economic_df_states['Rate Change from 2007 to 2008'],
          text = economic_df_states['Area_name'],
          colorbar = {'title': 'Î”% Rate Change from 2007 to 2008 (%)'},
          marker = dict(line = dict(color = 'rgb(255, 255, 255)', width=2)))
```

```python
# create a layout object with the dict() method

# scope arguemnts to the map of interest
# showlakes argument turns on display for lakes
# lakecolor adjusts color of lakes

layout = dict(title = 'Î”% Rate Change from 2007 to 2008 (%)',
            geo = dict(scope = 'usa', showlakes=True, lakecolor='rgb(85,173,240)'))
```

```python
# use the go.figure to set up the object that gets passed into iplot() method

choromap = go.Figure(data = [data], layout = layout)
```

```python
# use iplot() method to generate the plot
print(top_states_unemployed)
print(bottom_states_unemployed)
iplot(choromap)
```
