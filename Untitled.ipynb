```python
# File to Load 
gdp_data_to_load = "../Resources/GDP Per State.csv"

gdp_df = pd.read_csv(gdp_data_to_load)
gdp_df.dropna(inplace = True)

gdp_df.head()
```

```python
#Grab state name and industries
state_name = [state for state in gdp_df["GeoName"]]
industry_list = [indsutry for indsutry in gdp_df["Description"]]
```

```python
# Converting 2007 GDP
convert_2007 = []

for gdp in gdp_df["2007"]:
    if gdp=="(L)":
        convert_2007.append(0)
    else:
        convert_2007.append(float(gdp))
        
    
gdp_df["new_2007"]=convert_2007

# Grouping by state and sum it by state
grouped_gdp_2007 = gdp_df.groupby(["GeoName"])["new_2007"].sum()
grouped_gdp_2007
```

```python
# Create a dataframe
gdp_totals = pd.DataFrame({"GDP 2007": grouped_gdp_2007,
                            "GDP 2008": grouped_gdp_2008,
                            "GDP 2009": grouped_gdp_2009,
                            "GDP 2010": grouped_gdp_2010,
                            "GDP 2011": grouped_gdp_2011,
                            "GDP 2012": grouped_gdp_2012,
                            "GDP 2013": grouped_gdp_2013,
                            "GDP 2014": grouped_gdp_2014,
                            "GDP 2015": grouped_gdp_2015,
                            "GDP 2016": grouped_gdp_2016,
                            "GDP 2017": grouped_gdp_2017,
                            "GDP 2018": grouped_gdp_2018})
```

```python
# Calculate the delta from a year by year perspective for each state
delta_07_08 = ((gdp_totals["GDP 2008"] - gdp_totals["GDP 2007"])/ gdp_totals["GDP 2007"])*100
delta_08_09 = ((gdp_totals["GDP 2009"] - gdp_totals["GDP 2008"])/ gdp_totals["GDP 2008"])*100
delta_09_10 = ((gdp_totals["GDP 2010"] - gdp_totals["GDP 2009"])/ gdp_totals["GDP 2009"])*100
delta_10_11 = ((gdp_totals["GDP 2011"] - gdp_totals["GDP 2010"])/ gdp_totals["GDP 2010"])*100
delta_11_12 = ((gdp_totals["GDP 2012"] - gdp_totals["GDP 2011"])/ gdp_totals["GDP 2011"])*100
delta_12_13 = ((gdp_totals["GDP 2013"] - gdp_totals["GDP 2012"])/ gdp_totals["GDP 2012"])*100
delta_13_14 = ((gdp_totals["GDP 2014"] - gdp_totals["GDP 2013"])/ gdp_totals["GDP 2013"])*100
delta_14_15 = ((gdp_totals["GDP 2015"] - gdp_totals["GDP 2014"])/ gdp_totals["GDP 2014"])*100
delta_15_16 = ((gdp_totals["GDP 2016"] - gdp_totals["GDP 2015"])/ gdp_totals["GDP 2015"])*100
delta_16_17 = ((gdp_totals["GDP 2017"] - gdp_totals["GDP 2016"])/ gdp_totals["GDP 2016"])*100
delta_17_18 = ((gdp_totals["GDP 2018"] - gdp_totals["GDP 2017"])/ gdp_totals["GDP 2017"])*100
total_delta = ((gdp_totals["GDP 2018"] - gdp_totals["GDP 2007"])/ gdp_totals["GDP 2007"])*100
```

```python
gdp_complete = pd.DataFrame({"GDP 2007": grouped_gdp_2007, 
                             "GDP 2008": grouped_gdp_2008, 
                             "GDP 2009": grouped_gdp_2009, 
                             "GDP 2010": grouped_gdp_2010, 
                             "GDP 2011": grouped_gdp_2011, 
                             "GDP 2012": grouped_gdp_2012, 
                             "GDP 2013": grouped_gdp_2013, 
                             "GDP 2014": grouped_gdp_2014, 
                             "GDP 2015": grouped_gdp_2015, 
                             "GDP 2016": grouped_gdp_2016, 
                             "GDP 2017": grouped_gdp_2017, 
                             "GDP 2018": grouped_gdp_2018, 
                             "07-08 Delta": delta_07_08, 
                             "08-09 Delta": delta_08_09, 
                             "09-10 Delta": delta_09_10, 
                             "10-11 Delta": delta_10_11, 
                             "11-12 Delta": delta_11_12, 
                             "12-13 Delta": delta_12_13, 
                             "13-14 Delta": delta_13_14, 
                             "14-15 Delta": delta_14_15, 
                             "15-16 Delta": delta_15_16, 
                             "16-17 Delta": delta_16_17, 
                             "17-18 Delta": delta_17_18, 
                             "Total Delta": total_delta})

#gdp_complete.drop(["Far West", "Great Lakes", "Mideast", "New England", "Plains", 
                   #"Rocky Mountain", "Southeast", "Southwest"])
gdp_complete = gdp_complete.reset_index()
gdp_complete
```

```python
# Create dataframe to show 07-08 change and 08-18 change
gdp_change_07_08 = gdp_complete["09-10 Delta"]
gdp_change_08_18 = ((gdp_complete["GDP 2018"] - gdp_complete["GDP 2008"])/ gdp_complete["GDP 2008"])*100
state_name = [state for state in gdp_complete["GeoName"]]



gdp_for_visualization = pd.DataFrame({"State": state_name, 
                                      "GDP Change (2007-2008)": gdp_change_07_08, 
                                      "GDP Change (2008-2018)": gdp_change_08_18})


gdp_for_visualization
```

```python
print("Top performing states from 2007-2008")
top_performing_states_07_08 = gdp_for_visualization.sort_values(["GDP Change (2007-2008)"],ascending = False).head(5)
print(top_performing_states_07_08)

print("Bottom performing states from 2007-2008")
bottom_performing_states_07_08 = gdp_for_visualization.sort_values(["GDP Change (2007-2008)"],ascending = True).head(5)
print(bottom_performing_states_07_08)

print("Top performing states from 2008-2018")
top_performing_states_08_18 = gdp_for_visualization.sort_values(["GDP Change (2008-2018)"],ascending = False).head(5)
print(top_performing_states_08_18)

print("Bottom performing states from 2008-2018")
bottom_performing_states_08_18 = gdp_for_visualization.sort_values(["GDP Change (2008-2018)"],ascending = True).head(5)
print(bottom_performing_states_08_18)
```

```python
data = [go.Choropleth(
    locations = state_name,
    z = gdp_for_visualization["GDP Change (2007-2008)"],
    text = gdp_for_visualization["State"],
    colorscale = [
        [0, "rgb(5, 10, 172)"],
        [0.35, "rgb(40, 60, 190)"],
        [0.5, "rgb(70, 100, 245)"],
        [0.6, "rgb(90, 120, 245)"],
        [0.7, "rgb(106, 137, 247)"],
        [1, "rgb(220, 220, 220)"]
    ],
    autocolorscale = False,
    reversescale = True,
    marker = go.choropleth.Marker(
        line = go.choropleth.marker.Line(
            color = 'rgb(180,180,180)',
            width = 0.5
        )),
    colorbar = go.choropleth.ColorBar(
        tickprefix = '%',
        title = 'Rate'),
)]

layout = go.Layout(
    title = go.layout.Title(
        text = "GDP Change (2007-2008)"
    ),
    geo = go.layout.Geo(
        showframe = False,
        showcoastlines = False,
        projection = go.layout.geo.Projection(
            type = 'equirectangular'
        )
    ),
    annotations = [go.layout.Annotation(
        x = 0.55,
        y = 0.1,
        xref = 'paper',
        yref = 'paper',
        showarrow = False
    )]
)

fig = go.Figure(data = data, layout = layout)
py.iplot(fig, filename = 'd3-world-map')
```
